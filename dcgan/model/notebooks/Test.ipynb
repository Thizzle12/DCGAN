```python
import matplotlib.pyplot as plt
import torch
import torchvision

from dcgan.model.generator import Generator
```

```python
gen = Generator(latent_dims=256)

gen.load_state_dict(torch.load(r'./dcgan/model_dicts/generator_1.pt'))

```

```python
inputs = torch.randn(32, 100, 1, 1)
generated_img = gen(inputs)
print(generated_img.shape)

plt.figure()
plt.imshow(torch.permute(torch.squeeze(generated_img[0, ...]), (1, 2, 0)).detach().numpy())
plt.show()
# generated_img
```
