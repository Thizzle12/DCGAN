```python
import matplotlib.pyplot as plt
import torch

from dcgan.model.generator import Generator
```

```python
gen = Generator()

gen.load_state_dict(torch.load(r'./dcgan/model_dicts/generator_4.pt'))

inputs = torch.randn(1, 100, 1, 1)
generated_img = gen(inputs)
generated_img.shape
```

```python
plt.figure()
plt.imshow(torch.permute(torch.squeeze(generated_img), (1,2, 0)).detach().numpy())
plt.show()
```
